```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/cities.csv')
```

```python
df['population'].plot(kind='hist', bins=50, logy=True)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Distribution of Population')
plt.show()
```

```python
df['city_id'].plot(kind='hist', bins=50, logy=True)
df[df['population'] > 10000]['city_id'].plot(kind='hist', bins=50, logy=True)
plt.xlabel('City ID')
plt.ylabel('Frequency')
plt.title('Distribution of City IDs for Population > 10000')
plt.legend(['All entries', 'Population > 10000'])
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar')
top_countries_filtered = df[df['population'] > 10000]['country'].value_counts().head(10)
top_countries_filtered.plot(kind='bar', color='orange')
plt.xlabel('Country')
plt.ylabel('Number of Cities')
plt.title('Top 10 Countries with Most Occurring Cities')
plt.legend(['All entries', 'Population > 10000'])
plt.show()
```
